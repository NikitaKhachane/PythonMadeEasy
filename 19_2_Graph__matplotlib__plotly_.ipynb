```python
import cufflinks as cf
```

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

import matplotlib as plt
```

```python
cf.go_offline()
```

```python
ir = load_iris()

df = pd.DataFrame(data = ir.data, columns = ir.feature_names)
df['Species'] = ir.target_names[ir.target]
df
```

```python
df.iplot()  # now we get the graph that shoe values whenyou hower over it
# we can download the plot (picture)
# zoom in and zoom out
#drag and select a perticular part of graph to see it up close
```
