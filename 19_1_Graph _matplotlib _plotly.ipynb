```python
 # Boxplot

df.boxplot()  # boxplot : shows quantiles of the features
# used to find outliers
```

```python
df.describe()
```

```python
# we can see the 25%th, 50%th, 75%th percentiles in df.decribe()
# # these same percentiles can be seen marked in the boxplot
# it also helps us see the distribution of the data features

# For example : PetalLeangth has a lot of values in between 25th to 50 th percentile

# so,metimes there are data point that are outside the range of 9min,max) these points are called outliers


#     _____ -------------------(max)
#       |
#       |
#     __|__   -------Q3----------(75th %) percentile----------------
#    |     |                                                       |
#    |     |                                                       |
#    |     |                                                       |
#    |_____| --------Q2-----(50 % )  {Median}                     [ IQR ]
#    |     |                       : middle point of data          |
#    |     |                                                       |
#    |     |                                                       |
#    |     |                                                       |
#    |_____| -------Q1-----------(25 %) percentile------------------
#       |
#       |.    }\__  -----------[outliers]
#       |.    }/
#       |
#     __|__   -----------------(min)
#

# IQR = (Q3-Q1)

# min = Q1-1.5(IQR)

# max = Q3 - 1.5(IQR)
```

```python
# trying different variables
df.plot.scatter(x = "SepalLengthCm", y = "PetalWidthCm")
```

```python
df.plot.scatter(x= 'SepalLengthCm', y = 'PetalLengthCm', c = 'SepalWidthCm')
# what is the c doing :
# c is color column
# $ the range of color bar is 2-4.5
# color axis represents PetalLengthCm
# so basically :
# color intensity decides the SepalWidthCm value of the

# here the flowers that have high value for SepalWidthCm anre in lower corner i.e. have low 'SepalLengthCm' and'PetalLengthCm' (mostly)
# where as the flowers that have high 'SepalLengthCm'and 'PetalLengthCm' value have very low SepalWidthCm value and are in upper right corner
```

```python
# size of the dots in scatter plot
# s = number
# s : size parameter

df.plot.scatter(x= 'SepalLengthCm', y = 'PetalLengthCm', c = 'SepalWidthCm', s = 500)
```

```python
df.plot.scatter(x= 'SepalLengthCm', y = 'PetalLengthCm', c = 'SepalWidthCm', s = 300)
```

```python
# we can also contrl the size of the data instance with respect to a column
# for examople here "PetalWidthCm"

df.plot.scatter(x= 'SepalLengthCm', y = 'PetalLengthCm', c = 'SepalWidthCm', s = df["PetalWidthCm"]*500)
# so as the value of "PetalWidthCm" increses the dot size will also increse|
```
