```python
# to generate a cube:

def gencube(n):
  for i in range(n):
    return i**3
```

```python
gencube(6)
```

```python
def gencube_1(n):
  l = []
  for i in range(n):
    l.append(i**3)
  return l

gencube_1(4)
```

```python
def gen(n):
  for i in range(n):
    yield i**3

# yield = range like function and yield function only
# remember the following things

# 1) wht we produced last time,
# 2) whats the logic and
# 3) what where we have to stop
```

```python
gen(10)
```

```python
for i in gen(20):
  print(i)
# it is generating and thowing it in cosole
# gen() is a generator funtion
```

```python
# Fibbbonacci number sequences

# next number is sum of last 2 numbers

# 0  1  1 2 3 5 8 13..........
```

```python
# write a fun to prong above and create a gennerator not a list that has o hold all the informaiton
#  in memory
```

```python
def fib(n):
  a=1
  b =1
  for i in range(n):
    yield a, i
    a, b = b, a+b
```

```python
def feb1(n):
  a =0
  b = 1
  op = []
  for i in range(n):
    op.append(a)
    a,b = b , a+b
  return op

feb1(10)
```
