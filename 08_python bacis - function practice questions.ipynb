```python
#  1. you have to write a fun which will take string and return a len
#  of it without using a inbuilt fun len


def strlen(s):
  # docstring
  """this function accepts string output and returns the lenth of the string"""
  if type(s) == str:
    c=0
    for i in s:
      c+=1
    return c # return count
  else:
    return "TypeError : not avilable for ",type(s),"object type" # error msg

strlen("nikita")
```

```python
# 2. write a fun which will be able to print an index of all premitive
# element which you will pass


def prel(o):
  """prel(obj)
      return index of all Primitive element"""  # docstring
  k = []
  for i in range(len(o)):
    k.append([o[i],i])
  return k

prel([4,5,6,7])
```

```python
# 3. Write a fun which will take input as a dict and give me output as a
# list of all the values even in case of 2 level nesting it should work.

d = {"a" : 4555,
     "b" : "kite",
     "c" : [1,2,3,4],
     "d" : (9,8,"k","j"),
     "e" : 7.7,
     "f" : 5+5j}

def unravel(d):
  """unravel(obj)
     Return the values of dictionary \n
      Nested values will be unwraped and will be as seperate element in output
     """
  k = []
  if type(d) == dict:
    m = d.keys()
    for i in m:
      if type(d[i]) in (int, float, str, complex):
        k.append(d[i])
      elif type(d[i]) in (list, tuple):
         for j in d[i]:
          k.append(j)
      else:
        return "invalid input"
    return k
  else:
    return "Invalid Input obect"
```

```python
# 5. write a function whihc will take multiple list as a input and give
 # me concatnation of all the  and output

def con(*args):
  """con(*args)
  returns concatination of multiple lists"""
  l_all = []
  for i in args:
    if type(i) == list:
      l_all = l_all + i
    else:
      return l_all ,"Object type other than list excluded"
  return l_all

```

```python
l1 = [1,2,3,4,5,3]
l2 = ["sid","nik","pic"]
l3 = [45,3.4,9+5j]
l4 = [[4,5],"string",6.5]

t1 = (4,6,8,"nikee")
```
