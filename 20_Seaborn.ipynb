```python
# in same scatterplot plot smoer also

sns.relplot(x= 'total_bill', y = 'tip',data = df1, hue = 'smoker')  # hue is basiccaly groupbyoperation
```

```python
sns.relplot(x= 'total_bill', y = 'tip',data = df1, style = 'smoker')
```

```python
sns.relplot(x= 'total_bill', y = 'tip',data = df1, style = 'size')  # size of the group it means
```
